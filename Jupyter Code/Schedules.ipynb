```python
import json 
import csv

with open("C:/Users/Lakshya Agrawal/Downloads/archive/schedules.json") as f:
    data1 = json.load(f)
    
```

```python
with open("C:/Users/Lakshya Agrawal/Downloads/archive/Schedules.csv",'w') as file:
    csvfile = csv.writer(file)
    csvfile.writerow(['Arrival','Day','Train_name','Station_name','Station_code','Id','Train_number','Departure'])
    for data in data1:
        
        if(data['arrival']!=None ):
            csvfile.writerow([data['arrival'],data['day'],data['train_name'],data['station_name'],data['station_code'],data['id'],data['train_number'],data['departure']])
        
```
